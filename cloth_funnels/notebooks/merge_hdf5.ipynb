```python

import os
import h5py
import os


merge_h5py_files = [
    '/local/crv/acanberk/folding-unfolding/src/downstream_datasets/mixed-fling-prefold.hdf5',
    '/local/crv/acanberk/folding-unfolding/src/downstream_datasets/mixed-factorized-prefold.hdf5',
    '/local/crv/acanberk/folding-unfolding/src/downstream_datasets/mixed-unfactorized-prefold.hdf5'
]

out_file = '/local/crv/acanberk/folding-unfolding/src/downstream_datasets/mixed-all-prefold.hdf5'


```

```python

os.remove(out_file)
with h5py.File(out_file, 'w') as f_out:
    for path in merge_h5py_files:
        with h5py.File(path, 'r') as f:
            for k, v in f.items():
                print(k)
                new_key = k + str(hash(path))
                print(new_key)
                if new_key not in f_out:
                    f.copy(k, f_out, name=new_key, shallow=True)
```
