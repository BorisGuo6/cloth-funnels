```python
import numpy as np 
import imutils
import matplotlib.pyplot as plt 
import torchvision.transforms as transforms
import torch
```

```python
a = torch.randn(10, 128, 128)
```

```python
plt.imshow(a[:3, :, :].permute(1, 2, 0))
```

```python
zeros = torch.zeros_like(a)
scale = 3

a = transforms.functional.resize(a, (int(a.shape[-1]/scale), int(a.shape[-1]/scale)))
print("small a", a.shape) 

end = zeros.shape[-1]//2 + int(a.shape[-1])//2
begin = zeros.shape[-1]//2 - int(a.shape[-1])//2

if end - begin < a.shape[-1]:
    end += 1

zeros[..., begin:end, begin:end] = a 
plt.imshow(zeros[:3, :, :].permute(1, 2, 0))

print(begin, end)
```
